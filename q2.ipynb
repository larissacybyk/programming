```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1)

y = np.log(x)

z = np.exp(x)

plt.scatter(x, y, label='Natural Logarithm')
plt.ylabel("Natural Logarithm")
plt.xlabel("Numbers")
plt.title("Natural Log")
plt.legend(loc = 'lower right')

plt.show()

plt.scatter(x, z, label='Exponential Function')
plt.ylabel("Exponential Function")
plt.xlabel("Numbers")
plt.title("Exponential Functions")
plt.legend(loc = 'lower right')

plt.show()
```

```python
x = np.linspace(-6.5,6.5, num=100)

y = np.sin(x)
z = np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z)

plt.plot(x,y, label='sin')
plt.plot(x,z, label='cos')

plt.xlabel('Numbers')
plt.ylabel('Sine and Cosine')
plt.legend(loc='lower right')
plt.title('Sine and Cosine')

plt.show()
```
